```python
from llama_index.core import SimpleDirectoryReader, ServiceContext, VectorStoreIndex
# vector store index stores index of embedded vector
#Simledirectory reader reads data
#ServiceContext helps to combine llama to model with promt template
from llama_index.llms.huggingface import HuggingFaceLLM
# helps to interact with huggingface hubs
# from llama_index.prompts.prompts import SimpleInputPrompt
from llama_index.core.prompts.prompts import SimpleInputPrompt
from llama_index.core import PromptTemplate

#used to create prompts
```

```python
# reading documents
documents=SimpleDirectoryReader("/kaggle/input/data-resume-pdf").load_data()
documents
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""
## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

## Llama2 Model
- context window= size of the no of tokens can be given as prompt
- max_new_tokens= no of new tokens that can be generated
- temperature- measures 0 to 1 creativity of llm models
- query wrapper prompts- give prompt default setup
- tokenizer- llama2 tokenizer from its model wit 7 billion parameters
- model name - same model llama2 7 b model
- model_kwargs- to quantize and reduce model no of bits for gpu excecution
- datatye of model is float16


```python
import torch
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
    
```

## Embeddings
- This is a sentence-transformers model: It maps sentences & paragraphs to a 768 dimensional dense vector space and can be used for tasks like clustering or semantic search.

```python
# get embedings from langchain
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index.embeddings.langchain import LangchainEmbedding
embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

## Combinng all using service Context
- The service context container is a utility container for LlamaIndex index and query classes. The container contains the following objects that are commonly used for configuring every index and query, such as the LLM, the PromptHelper (for configuring input size/chunk size), the BaseEmbedding (for configuring the embedding model), and more.

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
# convert to vector from text
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
# create engine to querry
query_engine=index.as_query_engine()
```

```python
# store response
response=query_engine.query("what are names of candidates?")
```
